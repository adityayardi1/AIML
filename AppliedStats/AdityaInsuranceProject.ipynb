### Task 3.7: Checking the presence of outliers in â€˜bmiâ€™, â€˜ageâ€™ and â€˜charges columns:

###### Outliers: An outlier is a data point in a data set that is distant from all other observations. A data point that lies outside the overall distribution of the dataset.
###### Outliers cause serious issues for statistical analysis, skew the data, significant impact on mean and significant impact on standard deviation.
###### I am using IQR method to identify Outliers.


```python
insu_ol = insurance.drop(['children', 'sex', 'smoker', 'region'], axis = 1)
Q1 = insu_ol.quantile(0.25)
Q3 = insu_ol.quantile(0.75)
IQR = Q3 - Q1
print("Columnwise Inter Quratile Range ")
print(IQR)
print("---------------------------------------------------------")
print("---------------------------------------------------------")
print("Columnwise Lowerbound ")
lowerbound = (Q1 - 1.5 * IQR)
print(lowerbound)
print("---------------------------------------------------------")
print("Columnwise Upperbound ")

upperbound = (Q3 + 1.5 * IQR)
print(upperbound)

print("---------------------------------------------------------")
print("Determining Presence of Outliers: TRUE indicates presence of Outlier")
insu_outliers = ((insu_ol< (lowerbound)) | (insu_ol > (upperbound)))
print(insu_outliers)

print("---------------------------------------------------------")
print("Attribute wise Counts of Outliers")
print(insu_outliers['age'].value_counts())
print("---Thus, there are no outliers in age attribute---")
print(insu_outliers['bmi'].value_counts())
print(insu_outliers['charges'].value_counts())

print("---------------------------------------------------------")
print("Determining Outliers's values in Charges Attribute ")
insu_ol_charges = insu_ol[(insu_ol['charges'] > upperbound['charges']) | (insu_ol['charges'] < lowerbound['charges'])]['charges']
print( insu_ol_charges)

print("---------------------------------------------------------")
print("Determining Outliers's values in BMI Attribute ")
insu_ol_bmi = insu_ol[(insu_ol['bmi'] > upperbound['bmi']) | (insu_ol['bmi'] < lowerbound['bmi'])]['bmi']
print( insu_ol_bmi)
```

### Task 4: c) Is the proportion of smokers significantly different in different genders?


```python
# Finding total number of Smokers in the Dataset.
Number_of_Smokers = insurance[insurance["smoker"]== "yes"]["smoker"].count()
print(Number_of_Smokers)
```

```python
# Subsetting the dataset. Selecting only those observations of persons who smoke
insur = insurance[insurance["smoker"]== "yes"]
print(insur)
insur.drop(['age','bmi','children','region', 'charges'], axis=1, inplace=True)
print(insur)
Number_of_Female_Smokers = insur[insur["sex"]== "female"]["sex"].count()        # Counting number of Female Smokers
print("Number of Female Smokers =", Number_of_Female_Smokers)                 
Number_of_male_Smokers = insur[insur["sex"]== "male"]["sex"].count()            # Counting number of Male Smokers
print("Number of Male Smokers =", Number_of_male_Smokers)
```

##### plotting the categorical data: 'sex' and 'smokers' proportion using countplot. I am using countplot as it provides count of observations in each categorical bin using bars.

```python
total = float(len(insur)) 
ax = sns.countplot(x ="sex", data = insur) 
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
                            height + 3,
        '{0:.0%}'.format(height/total),
                           ha="center") 
```

##### From the above Statistical Analysis, it can be observed that out of 1338 persons, 274 are SMOKERS. Count of Male Smokers is: 159 which accounts for 58% of Smokers where as we have 115 Female Smokers in our dataset which contributes to 42 % of the total Smokers.    Thus, it can be concluded that the proportion of Smokers does vary on basis of Gender.
